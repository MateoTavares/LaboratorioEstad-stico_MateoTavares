1) Explica el modelo de regresión logística para clasificación. ¿Cómo se determina el umbral de decisión?

La regresión logística es un modelo de clasificación que te arroja probabilidades del 0 al 1, se ajusta a través de la máxima verosimilitud, es decir, una combinación parámetros dados tus datos X que maximicen la probabilidad de Y.
```python
import numpy as np
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF
from scipy.optimize import minimize
import warnings


warnings.filterwarnings('ignore')

def funcion_objetivo(params):
    x, y, z = params[0], params[1], params[2]

    x = np.clip(x, 0, 1)
    y = np.clip(y, 0, 1)
    z = np.clip(z, 0, 1)

    seno = np.sin(12*x - 4)
    seno2 = np.sin(12*z - 4)
    coseno = np.cos(12*y - 4)
    
    model = ((6*x - 2)**2) * seno + ((6*y - 2)**2) * coseno + ((6*z - 2)**2) * seno2
    
   
    return model
X_data = np.array([
    [0.1, 0.1, 0.1],
    [0.5, 0.5, 0.5],
    [0.9, 0.9, 0.9],
    [0.8, 0.8, 0.8],
    [0.7, 0.7, 0.7]
])


Y_data = np.array([funcion_objetivo(x) for x in X_data])

print("--- Puntos Iniciales ---")
print("X_data (Parámetros):")
print(X_data)
print("\nY_data (Resultados):")
print(Y_data)

kernel = RBF(length_scale=1.0, length_scale_bounds=(1e-1, 10.0))
regresion = GaussianProcessRegressor(kernel=kernel, 
                                     alpha=1e-6, 
                                     n_restarts_optimizer=9, 
                                     random_state=10)
n_iteraciones = 0
max_iteraciones = 30 
kappa = 1.96         

bounds = np.array([[0.0, 1.0], [0.0, 1.0], [0.0, 1.0]])

def funcion_adquisicion_lcb(X_test_point, gp_model, kappa):
    X_test_point = X_test_point.reshape(1, -1)
    mu, sigma = gp_model.predict(X_test_point, return_std=True)
    lcb = mu - kappa * sigma
    return lcb[0] 

# ====================================================================
# 4. CICLO PRINCIPAL DE OPTIMIZACIÓN BAYESIANA
# ====================================================================

print("\n--- INICIO DEL CICLO DE OPTIMIZACIÓN BAYESIANA ---")

while n_iteraciones < max_iteraciones:
    n_iteraciones += 1
    print(f"--- Iteración {n_iteraciones}/{max_iteraciones} ---")
    regresion.fit(X_data, Y_data)
    
    best_x_next = None
    best_lcb_value = np.inf
    n_restarts_optimizer = 5
    for _ in range(n_restarts_optimizer):
        x_start = np.random.uniform(bounds[:, 0], bounds[:, 1], bounds.shape[0])
        
        res = minimize(
            fun=funcion_adquisicion_lcb,
            x0=x_start,
            args=(regresion, kappa), 
            method='L-BFGS-B',
            bounds=bounds
        )
        
        if res.fun < best_lcb_value:
            best_lcb_value = res.fun
            best_x_next = res.x

    X_next = best_x_next
    
    if np.any(np.all(np.isclose(X_data, X_next, atol=1e-6), axis=1)):
        print(f"  > El optimizador sugiere un punto ya evaluado: {X_next}")
        print("  > Posible convergencia. Terminando optimización.")
    X_next_2d = X_next.reshape(1, -1) 
    Y_next_real = funcion_objetivo(X_next)
    
    print(f"  > Punto evaluado: {X_next}")
    print(f"  > Valor real: {Y_next_real:.4f} (LCB predicho: {best_lcb_value:.4f})")
    
    
    X_data = np.vstack([X_data, X_next_2d]) 
    Y_data = np.hstack([Y_data, Y_next_real]) 
  
    Y_min_real = np.min(Y_data)
    print(f"  > Mínimo real hasta ahora: {Y_min_real:.4f}")

# ====================================================================
#  FINAL DEL CICLO
# ====================================================================

print("\n!!! ÓPTIMIZACIÓN BAYESIANA COMPLETADA !!!")

idx_best_score = np.argmin(Y_data)
Y_best_real = Y_data[idx_best_score]
X_best_real = X_data[idx_best_score]

print(f"Se completaron {n_iteraciones} iteraciones.")
print("-" * 50)
print(f"El mínimo valor encontrado fue: {Y_best_real:.6f}")
print(f"Este valor se alcanzó con los parámetros X: {X_best_real}")
print("-" * 50)
```

```python
num_features = [
    'selling_price','original_price','reviews_count'
]
cat_features = ['availability','color','category','breadcrumbs'
]

target_column = 'review'
x = df[num_features + cat_features]
y = df[target_column]
```

```python
numerical_transformer=StandardScaler()
categorical_transformer = OneHotEncoder(handle_unknown='ignore')
preprocesser=ColumnTransformer(transformers=[
        ('num', numerical_transformer, num_features),
        ('cat', categorical_transformer, cat_features)
    ],)
```

```python
model= LogisticRegression(class_weight='balanced', max_iter=1000)
pipeline=Pipeline(steps=[
('preprocesser',preprocesser),
('model',model)])
modelo_final = pipeline.fit(X_train, y_train)
y_pred_proba = modelo_final.predict_proba(X_test)[:, 1]
auc_score_final = roc_auc_score(y_test, y_pred_proba)

```

```python
num_features = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age'
]
target_column = 'Outcome'
x = df1[num_features]
y = df1[target_column]
```

```python
numerical_transformer=StandardScaler()
preprocesser=ColumnTransformer(transformers=[
        ('num', numerical_transformer, num_features)])
```

```python
model= SVC(kernel='linear', class_weight='balanced',C=1.15,probability=True)
pipeline=Pipeline(steps=[
('preprocesser',preprocesser),
('model',model)])
```

```python

scores = cross_val_score(pipeline, x, y.values.ravel(), 
                               scoring='roc_auc_ovr', cv=10)
meanscore = np.mean(scores)
print(meanscore)
```

```python
def funcion_objetivo(X_hl) :
    model= SVC(kernel='linear', class_weight='balanced',C=X_hl,probability=True)
    pipeline=Pipeline(steps=[
    ('preprocesser',preprocesser),
    ('model',model)

    ])
    scores = cross_val_score(pipeline, x, y.values.ravel(), 
                               scoring='precision', cv=10)
    meanscore = np.mean(scores)
    return meanscore
```

```python
y1 = funcion_objetivo(0.001)
y2 = funcion_objetivo(1)
y3 = funcion_objetivo(10)
X_gauss= np.array([[.001],[1],[10]])
Y_scores = [y1, y2, y3]

kfolds = KFold(n_splits=10, shuffle=True, random_state=10)
Y_scores = np.array(Y_scores)
print("X_gauss =", X_gauss)
print("Y_scores =", Y_scores)
regresion = GaussianProcessRegressor(kernel=RBF(length_scale=1.0), random_state=10)
```

```python
# ====================================================================
# INICIO DEL CICLO DE OPTIMIZACIÓN BAYESIANA
# ====================================================================


max_ucb_value = 1.0 # Inicializamos el valor máximo en Y de los intervalos de confianza.
Y_max_real = Y_scores.max() # El mejor score real encontrado hasta ahora.
n_iteraciones = 0
max_iteraciones = 300 # Iteraciones del ciclo.

# 2. Preparar los arreglos para las nuevas observaciones
# Convertimos las listas iniciales a arreglos de numpy para facilitar la adición
X_data = np.array(X_gauss)
Y_data = Y_scores.copy() 

# --------------------------------------------------------------------
# CICLO PRINCIPAL
# --------------------------------------------------------------------

while (Y_max_real <= max_ucb_value) and (n_iteraciones < max_iteraciones):
    
    n_iteraciones += 1
    print(f"--- Iteración {n_iteraciones} ---")
    
    # A. AJUSTAR REGRESIÓN GAUSSIANA (Actualización del Modelo)
    # Se ajusta con TODOS los puntos (iniciales + nuevos)
    regresion.fit(X_data, Y_data)
    
    # Obtenemos la media y la varianza para los intervalos de confianza.
    n1_values = np.arange(0.001, 100, 0.01) 
    #n2_values = np.arange(5, 15, 1)
    #n1_grid, n2_grid = np.meshgrid(n1_values, n2_values)
    X_test_completo = n1_values.reshape([-1, 1])#np.c_[n1_grid.ravel(), n2_grid.ravel()]

    # FILTRADO CRÍTICO: Creamos una máscara para eliminar los puntos ya evaluados.
    # np.isin comprueba qué filas de X_test_completo YA están en X_data
    # ~ niega la máscara, seleccionando solo los que NO están en X_data
    is_not_evaluated = ~np.isin(X_test_completo, X_data, assume_unique=False).all(axis=1)
    
    # Los puntos candidatos reales a evaluar son solo los que no están en X_data
    X_test = X_test_completo[is_not_evaluated]

    # Si no quedan puntos para evaluar, terminamos el ciclo
    if len(X_test) == 0:
        print("No quedan puntos nuevos para explorar en X_test. Terminando.")
        break
    mu, sigma = regresion.predict(X_test, return_std=True)

    # C. CALCULAR UCB (intervalos de confianza)
    kappa = 1.96 
    ucb = mu + kappa * sigma
    
    # Obtener el valor UCB máximo
    max_ucb_value = np.max(ucb)
    print(f"Máximo UCB predicho en X_test: {max_ucb_value:.4f}")
    
    #Valor máximo de ucb (intervalos de confianza)
    idx_max_ucb = np.argmax(ucb)
    X_next = X_test[idx_max_ucb].reshape(1, -1)
    
    # E. EVALUAR EL PUNTO X_NEXT EN EL MUNDO REAL (Función Objetivo)
    Y_next_real = funcion_objetivo(X_next[0][0])
    
    print(f"  > Punto evaluado: {X_next[0]}, precicion real: {Y_next_real:.4f}")
    
    # F. ACTUALIZAR LOS ARRAYS DE DATOS Y EL MÁXIMO REAL
    X_data = np.vstack([X_data, X_next]) # Añadir la X al set de entrenamiento
    Y_data = np.hstack([Y_data, Y_next_real]) # Añadir la Y al set de entrenamiento
    
    # Actualizar el máximo ROC AUC real encontrado hasta ahora
    Y_max_real = np.max(Y_data) 
    print(f"  > Máxima precision real hasta ahora: {Y_max_real:.4f}")
    
    print("-" * 50)
    
# ====================================================================
# FINAL DEL CICLO
# ====================================================================

print("\n!!! ÓPTIMIZACIÓN BAYESIANA COMPLETADA !!!")
print(f"La condición de parada se cumplió en la iteración {n_iteraciones}.")
print(f"Precision máxima encontrada es: {Y_max_real:.4f}")
idx_best_score = np.argmax(Y_data)
Y_best_real = Y_data[idx_best_score]
X_best_real = X_data[idx_best_score]
print(f"Este valor se alcanzó con los parámetros X: {X_best_real}")
```
