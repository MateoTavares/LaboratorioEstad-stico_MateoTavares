## A07 Árboles de regresión



```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
from sklearn.linear_model import Ridge
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
```

```python
df=pd.read_csv('Advertising.csv')
```

```python
df.head()
```

```python
X = df.drop(columns=["sales"])
Y = df["sales"]
X_entrenamiento, X_prueba, y_entrenamiento, y_prueba = train_test_split(
    X, Y, test_size=0.6, random_state=42
)

scaler = StandardScaler()
X_entrenamiento_esc = scaler.fit_transform(X_entrenamiento)
X_prueba_esc = scaler.transform(X_prueba)

modelo = LinearRegression()
modelo.fit(X_entrenamiento_esc, y_entrenamiento)

r2_train = modelo.score(X_entrenamiento_esc, y_entrenamiento)
r2_test = modelo.score(X_prueba_esc, y_prueba)

print("R2 Train:", r2_train)
print("R2 Test:", r2_test)
```

```python
model_dt=DecisionTreeRegressor()
model_dt.fit(X_entrenamiento_esc, y_entrenamiento)
r2_traindt = model_dt.score(X_entrenamiento_esc, y_entrenamiento)
r2_testdt = model_dt.score(X_prueba_esc, y_prueba)

print("R2 Train:", r2_traindt)
print("R2 Test:", r2_testdt)
```

__ Poli

```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
```

```python
df=pd.read_csv('Advertising.csv')
X = df.drop(columns=["sales"])
Y = df["sales"]

X_entrenamiento, X_prueba, y_entrenamiento, y_prueba = train_test_split(
    X, Y, test_size=0.6, random_state=42
)
poly = PolynomialFeatures(degree=2)
modelo = LinearRegression()
model_dt=DecisionTreeRegressor()

poly_reg = Pipeline([('scaler',scaler),
    ('poly', poly),
    ('lin_reg', modelo)
])


poly_tree = Pipeline([('scaler',scaler),
    ('poly', poly),
    ('tree_reg', model_dt)
])

deg1_reg=poly_reg.fit(X_entrenamiento, y_entrenamiento)
deg1_tree=poly_tree.fit(X_entrenamiento, y_entrenamiento)
print('Regresion')
r2_train = deg1_reg.score(X_entrenamiento, y_entrenamiento)
r2_test = deg1_reg.score(X_prueba, y_prueba)

print("R2 Train:", r2_train)
print("R2 Test:", r2_test)

print('Arbol')
r2_train_a = deg1_tree.score(X_entrenamiento, y_entrenamiento)
r2_test_a = deg1_tree.score(X_prueba, y_prueba)

print("R2 Train:", r2_train_a)
print("R2 Test:", r2_test_a)
```
