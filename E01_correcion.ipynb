3.        ¿En qué consiste el proceso de escalamiento de factores?
Poner en la misma escala los datos de X, de forma que los coeficientes puedan ser analizados por impacto. Este se hace normalmente de forma estándar, es decir restando la media y dividiendo por la desviación estándar.

4.        Explica el propósito de penalizar factores en una regresión.
Penalizar factores consiste en no permitir que coeficientes no tengan demasiado peso en el modelo, ya que esto puede llevar al overfitting, con las penalizaciones, puedes reducir coeficientes demasiado alejados o también eliminar los que no son significativos.

5.        ¿Cuál es la relación entre escalamiento y penalización?
La penalización es sobre los coeficientes, por tanto si no están en la misma escala, la penalización no podrá identificar si un coeficiente tiene demasiado o muy poco peso.

## Ejercicio 3 práctico corregido.

```python
df = pd.read_csv('imdb_top1000_lae.csv')
df.head()
```

```python
df= df.drop(columns = ['Unnamed: 0'])
df= df.drop(columns = ['Director'])
```

Eliminé director porque eran demasiadas columnas extra.

```python

df["Runtime"] = df["Runtime"].str.replace(" min", "").astype(int)
```

```python
df_con_dummies = pd.get_dummies(df, columns=['Certificate'])
```

```python
bool_cols = df_con_dummies.select_dtypes(include=bool).columns
df_con_dummies[bool_cols] = df_con_dummies[bool_cols].astype(int)
```

```python
df_con_dummies
```

```python
df_con_dummies["Gross"] = df_con_dummies["Gross"].replace(",", "", regex=True).astype(float)
df_con_dummies["Released_Year"] = pd.to_numeric(df_con_dummies["Released_Year"], errors="coerce")
```

```python
df_con_dummies= df_con_dummies.dropna()
```

```python
from scipy import stats
X = df_con_dummies.drop(columns=["Series_Title", "IMDB_Rating"])
Y = df_con_dummies["IMDB_Rating"]

# Train-Test Split
X_entrenamiento, X_prueba, y_entrenamiento, y_prueba = train_test_split(
    X, Y, test_size=0.6, random_state=137
)

# Escalar
scaler = StandardScaler()
X_entrenamiento_esc = scaler.fit_transform(X_entrenamiento)
X_prueba_esc = scaler.transform(X_prueba)

# Modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(X_entrenamiento_esc, y_entrenamiento)

# R2 en train y test
r2_train = modelo.score(X_entrenamiento_esc, y_entrenamiento)
r2_test = modelo.score(X_prueba_esc, y_prueba)

print("R2 Train:", r2_train)
print("R2 Test:", r2_test)
```

R2 demasiado bajo, existen muchas más variables, más complejas que determinan la calificación de una película.

```python

n = X_entrenamiento_esc.shape[0]
p = X_entrenamiento_esc.shape[1] + 1   
X_train_const = np.hstack([np.ones((n, 1)), X_entrenamiento_esc])
y_train = y_entrenamiento.values


XtX = X_train_const.T @ X_train_const
XtX_pinv = np.linalg.pinv(XtX)
beta_hat = XtX_pinv @ (X_train_const.T @ y_train)


y_hat = X_train_const @ beta_hat
resid = y_train - y_hat

rank_X = np.linalg.matrix_rank(X_train_const)
df_res = n - rank_X
RSS = float(resid @ resid)
sigma2 = RSS / df_res

var_beta = sigma2 * XtX_pinv
se_beta = np.sqrt(np.diag(var_beta))


t_stats = beta_hat / se_beta
p_values = 2 * (1 - stats.t.cdf(np.abs(t_stats), df=df_res))


feature_names = ["Intercepto"] + list(X.columns)
res_df = pd.DataFrame({
    "Termino": feature_names,
    "p_value": p_values
})
res_df.head(34)
```

Vemos que casi todas las clasificaciones y géneros son insignificantes, salvo el género western, la duración y el metascore.
